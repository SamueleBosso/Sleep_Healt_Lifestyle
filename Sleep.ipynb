# 2. **Is there a relationship between age and the frequency of sleep disorders?**

```python
#let's start with the creation of differents age groups
Childhood = sleep[(sleep['Age'] >= 0) & (sleep['Age'] <= 14)].shape[0]
Young_Adults = sleep[(sleep['Age'] >= 14) & (sleep['Age'] <= 24)].shape[0]
Middle_Aged_Adults = sleep[(sleep['Age'] >= 24) & (sleep['Age'] <= 44)].shape[0]
Mature_Adults = sleep[(sleep['Age'] >= 44) & (sleep['Age'] <= 64)].shape[0]
Ancient = sleep[sleep['Age'] >= 65].shape[0]
```

```python
# creation of the new coloumn called age_group
def age_group(age):
    if 0 <= age <= 14:
        return 'Childhood'
    elif 14 < age <= 24:
        return 'Young Adults'
    elif 24 < age <= 44:
        return 'Middle Aged Adults'
    elif 44 < age <= 64:
        return 'Mature Adults'
    else:
        return 'Ancient'
    
```

```python
# we make union of age group in the dataset
sleep['Age Group'] = sleep['Age'].apply(age_group)
sleep.head()
```

```python
# Grouping the data by Age Group and Sleep Disorder, counting occurrences, and unstacking for a tabular format
counts = sleep.groupby(['Age Group', 'Sleep Disorder']).size().unstack(fill_value=0)
```

```python
# Plotting the frequency of sleep disorders by age group as a stacked bar chart
counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age Group')  # Label for the X-axis
plt.ylabel('Frequency')  # Label for the Y-axis
plt.title('Frequency of Sleep Disorders by Age Group')  # Title of the chart
plt.legend(title='Sleep Disorder')  # Legend with a title
plt.tight_layout()  # Adjust layout for better visualization
plt.show()
```
